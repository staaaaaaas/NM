```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [-1.0, 0.0, 1.0, 2.0, 3.0]
y = [-0.5, 0.0, 0.5, 0.86603, 1.0]
x_ans = 1.0

plt.plot(x,y)
plt.scatter(x_ans,y[2])
```

```python
# посмотрим, куда попадает точка x_ans
def find_start(x, p):
    for i in range(0, len(p) - 1):
        if p[i] <= x and x <= p[i + 1]:
            return i
```

```python
find_start(x_ans, x)
# точка x_ans попадает в отрезок [x1, x2], совпадает с правой границей отрезка
# ищем со вторым порядком точности 
```

```python
def df1(x, y, x0):
    i = find_start(x0, x)
    elem1 =  (y[i + 1] - y[i]) / (x[i + 1] - x[i])
    elem2 =  ((y[i + 2] - y[i + 1]) / (x[i + 2] - x[i + 1]) - elem1) / (x[i + 2] - x[i]) * (2 * x0 - x[i] - x[i + 1])
    return elem1 + elem2

def df2(x, y, x0):
    i = find_start(x0, x)
    elem1 = (y[i + 2] - y[i + 1]) / (x[i + 2] - x[i + 1])
    elem2 = (y[i + 1] - y[i]) / (x[i + 1] - x[i])
    return 2 * (elem1 - elem2) / (x[i + 2] - x[i])
```

```python
df1(x,y,x_ans)
```

```python
df2(x, y, x_ans)
```
